```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Q.1 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find 
# i. Sum of all elements 
# ii. Sum of all elements row-wise 
# iii. Sum of all elements column-wise 
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
sum_gfg = np.sum(gfg)
sum_row_wise = np.sum(gfg, axis = 1)
sum_col_wise = np.sum(gfg, axis = 0)
print("Sum of all elements: ",sum_gfg)
print("Sum of all elements row-wise: \n" ,sum_row_wise)
print("Sum of all elements column-wise: \n" ,sum_col_wise)
```

```python
# Q.2 (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find 
# i. Sorted array 
# ii. Indices of sorted array 
# iii. 4 smallest elements 
# iv. 5 largest elements 
# (b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find 
# i. Integer elements only 
# ii. Float elements only 
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
indices_sorted = np.argsort(array)
smallest_elements = np.partition(array, 4)[:4]
largest_elements = np.partition(array, -5)[-5:]

print("i. Sorted array:", sorted_array)
print("ii. Indices of sorted array:", indices_sorted)
print("iii. 4 smallest elements:", smallest_elements)
print("iv. 5 largest elements:", largest_elements)
```

```python
array_float = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array_float[array_float.astype(int) == array_float]
float_elements = array_float[array_float.astype(int) != array_float]

print("i. Integer elements only:", integer_elements)
print("ii. Float elements only:", float_elements)
```

```python
#Q3. You are given a weekly sales dataset and need to perform various operations 
# using NumPy broadcasting. 
# a) Generate your unique sales dataset: 
#  Take the sum of the ASCII values of the initials of your first and last 
# name. Call this value X. (If your initials are A B → ASCII sum = 65 
# + 66 = 131 → sales = [131, 181, 231, 281, 331].) 
#  Create a NumPy array sales with values [X, X+50, X+100, X+150, 
# X+200]. 
# b) Compute your personalized tax rate as ((X % 5) + 5) / 100. 
#  Use broadcasting to apply this tax rate to each sales value. 
# c) Adjust sales based on discount: 
#  If sales < X+100, apply a 5% discount. 
#  If sales >= X+100, apply a 10% discount. 
# d) Expand sales data for multiple weeks: 
#  Create a 3×5 matrix representing three weeks of sales by stacking 
# sales three times using broadcasting. 
#  Increase sales by 2% per week using element-wise broadcasting.
```

```python
initials = 'BJ'  # Example initials
X = sum(ord(char) for char in initials)
sales = np.array([X, X + 50, X + 100, X + 150, X + 200])

tax_rate = ((X % 5) + 5) / 100
sales_with_tax = sales * (1 + tax_rate)

discounted_sales = np.where(sales < X + 100, sales_with_tax * 0.95, sales_with_tax * 0.90)

weekly_sales = np.tile(discounted_sales, (3, 1))  
weekly_sales *= (1 + 0.02 * np.arange(3)[:, None])

print("Weekly Sales Data:\n", weekly_sales)
```

```python
# Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use 
# each function from the list below and compute y values using NumPy: 
#  Y = x2 
#  Y = sin(x) 
#  Y = e^x 
#  Y = log(|x| + 1) 
# Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity. 
```

```python
x = np.linspace(-10,10,100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x)+1)

plt.figure(figsize=(5, 3))
plt.plot(x, y1, label='Y = x²')
plt.title('Plot of Y = x²')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, y2, label = 'Y = sinx')
plt.title('Plot of Y = sinx')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y3, label='Y = e^x')
plt.title('Plot of Y = e^x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y4, label='Y = log(|x| + 1)')
plt.title('Plot of Y = log(|x| + 1)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid()
plt.legend()
plt.show()
```
