```python
# Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
#  Generate x values from -10 to 10 using np.linspace().
#  Compute y values for:
# o y= M⋅x
# 2
# o y=M⋅sin(x)
#  Plot both functions in a single figure:
# o Use different colors and line styles.
# o Add legend, grid, and title. 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
```

```python
M = float(input("Enter a value for M: "))
```

```python
x = np.linspace(-10,10,100)
```

```python
y1 = M * x**2
y2 = M * np.sin(x)
```

```python
plt.plot(x, y1, label = 'y = M * x^2', color = 'blue', linestyle = '-')
plt.plot(x, y2, label = 'y =  M * sin(x)', color = 'green', linestyle = '--')

plt.legend()
plt.grid(True)
plt.title('Graphs of the functions')

plt.show()
```

```python
# Q.2 Create a dataset of five subjects and scores.
#  Convert it into a Pandas DataFrame.
#  Plot the scores using a Seaborn bar plot with:
# o Different colors for each bar.
# o Annotations on top of each bar.
# o Title, axis labels, and grid. 
```

```python
data = {'Subject' : ['Maths', 'Physics', 'English', 'History', 'Geography'],
        'Scores'  :  [random.randint(60, 100) for _ in range(len(['Maths', 'Physics', 'English', 'History', 'Geography']))]
       }

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Subject', y='Scores', data=df, hue='Subject', palette='viridis', legend=False)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                xytext=(0, 5), textcoords='offset points',
                ha='center', va='center', fontsize=10, color='black')

plt.title('Student Scores in Different Subjects')
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.grid(True)

plt.show()
```

```python
# Q3. Write a Python script to select your roll number as seed for NumPy and
# then generate a dataset of 50 values using np.random.randn().
#  Create a 2x2 subplot layout displaying:
# o Line plot showing cumulative sum.
# o Scatter plot with random noise.
#  Customize with titles, axis labels, and grids. 
```

```python

np.random.seed(102317149)

dataset = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(dataset))
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].set_xlabel('Data Points')
axes[0, 0].set_ylabel('Cumulative Sum')
axes[0, 0].grid(True)


noise = np.random.randn(50)  
axes[0, 1].scatter(dataset, dataset + noise)
axes[0, 1].set_title('Scatter Plot with Random Noise')
axes[0, 1].set_xlabel('Dataset')
axes[0, 1].set_ylabel('Dataset + Noise')
axes[0, 1].grid(True)

axes[1, 0].plot(dataset) 
axes[1, 0].set_title('Line Plot')
axes[1, 0].set_xlabel('Data Points')
axes[1, 0].set_ylabel('Dataset Values')
axes[1, 0].grid(True)

# Lower right subplot (scatter plot)
axes[1, 1].scatter(np.arange(50), dataset) 
axes[1, 1].set_title('Scatter Plot')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('Dataset Values')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Q.4 Download Data-set from the below link
# https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
# Apply ‘seaborn’ library to do the following.
# 1. Read Total profit of all months and show it using a line plot.
# 2. Read all product sales data and show it using a multiline plot.
# 3. Plot bar chart for all the features/attributes. 
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
plt.figure(figsize = (10,6))
sns.lineplot(x = 'month_number', y = 'total_profit', data = df)
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10,6))
for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    sns.lineplot(x = 'month_number', y = product, data = df, label = product)
plt.title('Sales Data of All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (12,8))
for feature in df.columns[1:]:
    plt.figure() #new figure for each feature
    sns.barplot(x = df.index, y = feature, data =df)
    plt.xlabel('Index')
    plt.ylabel(feature)
    plt.xticks(rotation = 45)
    # plt.grid(True)
    plt.show()
```
